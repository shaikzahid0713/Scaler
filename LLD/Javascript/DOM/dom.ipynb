### Event Listener

Sure! Here's a list of some common and useful event listeners in JavaScript that you can use with DOM elements:

#### 1. **Mouse Events:**
- **`click`**: Triggered when an element is clicked.
- **`dblclick`**: Triggered when an element is double-clicked.
- **`mousedown`**: Triggered when the mouse button is pressed down on an element.
- **`mouseup`**: Triggered when the mouse button is released over an element.
- **`mousemove`**: Triggered when the mouse pointer moves over an element.
- **`mouseover`**: Triggered when the mouse pointer enters an element.
- **`mouseout`**: Triggered when the mouse pointer leaves an element.
- **`mouseenter`**: Triggered when the mouse pointer enters an element (doesnt bubble).
- **`mouseleave`**: Triggered when the mouse pointer leaves an element (doesnt bubble).

#### 2. **Keyboard Events:**
- **`keydown`**: Triggered when a key is pressed down.
- **`keyup`**: Triggered when a key is released.
- **`keypress`**: Triggered when a key is pressed and a character is produced (deprecated, use `keydown` instead).

#### 3. **Focus Events:**
- **`focus`**: Triggered when an element gains focus (e.g., an input field).
- **`blur`**: Triggered when an element loses focus (e.g., when clicking out of an input field).
- **`focusin`**: Triggered when an element or any of its descendants receives focus (bubbles).
- **`focusout`**: Triggered when an element or any of its descendants loses focus (bubbles).

#### 4. **Form Events:**
- **`submit`**: Triggered when a form is submitted.
- **`change`**: Triggered when an input elements value changes (for `<input>`, `<select>`, `<textarea>`).
- **`input`**: Triggered when the value of an `<input>`, `<textarea>`, or `<select>` element changes (works for real-time changes).
- **`reset`**: Triggered when a form is reset.
- **`select`**: Triggered when text is selected inside an input or textarea element.

#### 5. **Window Events:**
- **`resize`**: Triggered when the window is resized.
- **`scroll`**: Triggered when the document or an element is scrolled.
- **`load`**: Triggered when the document, image, or other media is fully loaded.
- **`beforeunload`**: Triggered before the window is unloaded (before closing the page).
- **`unload`**: Triggered when the page is unloaded (e.g., closing the browser tab).
- **`orientationchange`**: Triggered when the orientation of the device changes (portrait to landscape or vice versa).

#### 6. **Touch Events (for mobile devices):**
- **`touchstart`**: Triggered when the user touches the screen.
- **`touchmove`**: Triggered when the user moves their finger on the screen.
- **`touchend`**: Triggered when the user lifts their finger off the screen.
- **`touchcancel`**: Triggered when the touch event is interrupted (e.g., by an incoming call).

#### 7. **Pointer Events:**
- **`pointerdown`**: Triggered when a pointer (mouse, touch, stylus) is pressed.
- **`pointerup`**: Triggered when a pointer is released.
- **`pointermove`**: Triggered when a pointer is moved.
- **`pointerenter`**: Triggered when a pointer enters an element.
- **`pointerleave`**: Triggered when a pointer leaves an element.
- **`pointercancel`**: Triggered when a pointer is canceled.

#### 8. **Clipboard Events:**
- **`copy`**: Triggered when the user copies content.
- **`cut`**: Triggered when the user cuts content.
- **`paste`**: Triggered when the user pastes content.

#### 9. **Media Events:**
- **`play`**: Triggered when media (audio/video) starts playing.
- **`pause`**: Triggered when media (audio/video) is paused.
- **`ended`**: Triggered when media (audio/video) ends.
- **`volumechange`**: Triggered when the volume of media is changed.
- **`timeupdate`**: Triggered when the current time of media changes.
- **`loadeddata`**: Triggered when media data is loaded.

#### 10. **Animation Events:**
- **`animationstart`**: Triggered when a CSS animation starts.
- **`animationend`**: Triggered when a CSS animation ends.
- **`animationiteration`**: Triggered when a CSS animation repeats.

#### 11. **Transition Events:**
- **`transitionend`**: Triggered when a CSS transition finishes.

#### 12. **Other Events:**
- **`error`**: Triggered when an error occurs while loading a resource (like an image).
- **`dragstart`**: Triggered when an element starts being dragged.
- **`drag`**: Triggered when an element is being dragged.
- **`dragend`**: Triggered when a dragged element finishes dragging.
- **`dragenter`**: Triggered when a dragged element enters a drop target.
- **`dragleave`**: Triggered when a dragged element leaves a drop target.
- **`dragover`**: Triggered when a dragged element is over a drop target.
- **`drop`**: Triggered when a dragged element is dropped on a target.

---

#### Summary:
These are some of the most common event types in JavaScript. There are many other specialized events, but these should cover most use cases you'll encounter while working with the DOM. You can add event listeners for any of these events using `addEventListener()`.
